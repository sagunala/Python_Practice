```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [15,25,48,78,96,45,78]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='g',linewidth=1,marker ='D',alpha=0.5)
```

```python
x = [1,2,3,4,5,6,7]
max = [15,25,35,45,65,95,85]
min=[12,26,31,54,66,44,19]
avg = [25,35,45,55,65,75,15]

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")


plt.plot(x,max,label= "Max")
plt.plot(x,min,label= 'Min')
plt.plot(x,avg,label='Avg')

plt.legend(loc = 'best', fontsize = 'large', shadow=True)
plt.grid()
```

```python
import numpy as np
```

```python
names = ['Ram','Krish','Gan','Laxm']
marks = [40,55,70,90]
```

```python
xpos = np.arange(len(names))
xpos
```

```python
plt.bar(xpos,marks, label='marks')
```

```python
plt.bar(xpos,marks, label='marks')
plt.xticks(xpos, names)
plt.ylabel("marks")
plt.title("Class Marks")
plt.legend()
```

```python
Grades = ['A','B','C','D']
plt.bar(xpos-0.5, marks, label='marks')
plt.bar(xpos+0.5, Grades, label='Grades')
plt.xticks(xpos, names)
plt.ylabel('marks')
plt.title('Class Marks')
plt.legend()
```

```python
plt.barh(xpos-0.5, marks, label='marks')
plt.barh(xpos+0.5, Grades, label='Grades')
plt.xticks(xpos, names)
plt.ylabel('marks')
plt.title('Class Marks')
plt.legend()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.7)
```

```python
plt.hist(blood_sugar, rwidth=0.8, bins=5)
```

```python
plt.xlabel("Sugar level")
plt.ylabel("Patients")
plt.title("Patients Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins = [80,100,125,150], rwidth=0.8, label=["men","women"], color=["g", "orange"])
plt.legend()
```

```python
plt.hist([blood_sugar_men,blood_sugar_women], bins = [80,100,125,150], rwidth=0.8, label=["men","women"], color=["g", "orange"],orientation='horizontal')
plt.legend()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x= [1400,600,300,410,250]
y = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(x,labels=y, radius=2,autopct = '%0.1f%%', explode =[0,0.1,0,0,0])
plt.show()
plt.savefig("chart.pdf",bbox_inches='tight', transparent=True,pad_inches=1)
```
